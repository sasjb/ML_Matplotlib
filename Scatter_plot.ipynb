```python
df_google_playstore_app = pd.read_csv("D:\\New folder\googleplaystore.csv", nrows=1000)
df_google_playstore_app.shape
```

```python
df_google_playstore_app
```

```python
x= df_google_playstore_app["Rating"]
y = df_google_playstore_app["Reviews"]
plt.scatter(x,y)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c='r',marker = "*", alpha = 0.3)
plt.title("Google Playstore Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (10,9))
plt.scatter(x,y,c='r',marker = "*",s=50, alpha = 0.3,linewidth=5, edgecolor="g")
plt.scatter(x,df_google_playstore_app["Installs"],c='y',marker = "o", s=50, alpha = 0.3,linewidth=5, edgecolor="c")
plt.title("Google Playstore Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews and Installs")
```
