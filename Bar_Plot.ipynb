Bar Char Plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from matplotlib import style
```

```python
classes = ["python", "C", "C++", "Java", "Algorithm"]
class_1 = [10,27,15,15,33]
class_2 = [15,25,17,23,20]
class_3 = [28,2,50,12,8]
```

```python
plt.bar(classes,class_1)
```

```python
plt.bar(classes,class_1, width = 0.6,color = 'y', align = "edge" , edgecolor = "blue", linewidth = 4, alpha = 0.6, linestyle = "--")
```

```python
style.use("ggplot")
plt.figure(figsize= (16,9))
plt.bar(classes,class_1, width = 0.6,color = 'y', align = "edge" , edgecolor = "blue", linewidth = 4, alpha = 0.6, linestyle = "--")
```

```python
plt.figure(figsize= (16,9))
plt.bar(classes,class_1, width = 0.6,color = 'y', align = "edge" , edgecolor = "blue", linewidth = 4, alpha = 0.6, linestyle = "--")
plt.title("Bar Chart of CSE Students", fontsize = 18)
plt.xlabel("Number of Subject")
plt.ylabel("Number of Student")
```

```python
class_index = np.arange(len(classes))
width = 0.2
plt.bar(class_index,class_1, width,color = 'y')
plt.bar(class_index+width,class_2, width,color = 'b')
plt.bar(class_index+width+width,class_3, width,color = 'r')
plt.xticks(class_index+width,classes,rotation = 20)
plt.title("Bar Chart of CSE Students", fontsize = 18)
plt.xlabel("Number of Subject")
plt.ylabel("Number of Student")
```
