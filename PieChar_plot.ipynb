Pie Char Plot

```python
import matplotlib.pyplot as plt
```

```python
classes = ["python", "C", "C++", "Java", "Algorithm"]
class_1 = [10,27,15,15,33]
```

```python
plt.pie([1])
```

```python
plt.pie(class_1,labels = classes)
```

```python
classes = ["python", "C", "C++", "Java", "Algorithm"]
class_1 = [10,27,15,15,33]
explode = [0.1,0.1,0.1,0,0]
plt.pie(class_1,labels = classes,explode=explode)
```

```python
classes = ["python", "C", "C++", "Java", "Algorithm"]
class_1 = [10,27,15,15,33]
explode = [0.1,0.1,0.1,0,0]
colors = ['g','b','y','r','c']
plt.pie(class_1,labels = classes,explode=explode,colors = colors, autopct= "%0.2f%%")
```

```python
explode = [0.1,0,0.2,0,0]
colors = ['g','b','y','r','c']
textprops = {"fontsize":15}
plt.pie(class_1,labels = classes,explode=explode,colors = colors, autopct= "%0.2f%%", shadow = True,radius=1.4,textprops = textprops)
```

```python
plt.figure(figsize= (8,4))
wedgeprops = {"linewidth":4, "width":3, "edgecolor":"k"}
explode = [0.1,0,0.2,0,0]
colors = ['g','b','y','r','c']
textprops = {"fontsize":15}
plt.pie(class_1,labels = classes,explode=explode,colors = colors, autopct= "%0.2f%%", shadow = True,radius=0.8,textprops = textprops,
        wedgeprops = wedgeprops,frame = True, rotatelabels = True)
```

```python
explode = [0.1,0,0.2,0,0]
colors = ['g','b','y','r','c']
textprops = {"fontsize":15}
plt.pie(class_1,labels = classes,explode=explode,colors = colors, autopct= "%0.2f%%", shadow = True,radius=1.4,textprops = textprops)
plt.legend()
```
