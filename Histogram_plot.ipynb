Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
```

```python
print(py_student_age)
```

```python
plt.hist(ml_student_age, rwidth = 0.8)
plt.title("ML Students age Histogram")
plt.xlabel("Student age category")
plt.ylabel("No. of Students")
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8)
plt.title("ML Students age Histogram")
plt.xlabel("Student age category")
plt.ylabel("No. of Students")
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, orientation = "vertical" , color = "r", label = "ML Students")
plt.title("ML Students age Histogram")
plt.xlabel("Student age category")
plt.legend()
plt.ylabel("No. of Students")
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age], bins, rwidth = 0.8, orientation = "vertical" , color = ["r","y"], label = ["ML Students","Py Students"])
plt.title("ML Students age Histogram")
plt.xlabel("Student age category")
plt.legend()
plt.ylabel("No. of Students")
```

```python
from matplotlib import style
bins = [15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age], bins, rwidth = 0.8, orientation = "vertical" , color = ["r","y"], label = ["ML Students","Py Students"])
plt.title("ML Students age Histogram")
plt.xlabel("Student age category")
style.use("ggplot")
plt.legend()
plt.ylabel("No. of Students")
```
